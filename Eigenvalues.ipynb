```python
import numpy as np
import matplotlib.pyplot as plt


print("Vaishnavi Karma")
print("B.Tech VII sem")
print("Roll No. 0873CS211115")

# Sample matrix
A = np.array([[2, 1], [1, 3]])

# Performing eigenvalue decomposition
eigenvalues, eigenvectors = np.linalg.eig(A)

# Visualizing the results
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Plotting the eigenvectors (assuming 2D)
plt.arrow(0, 0, eigenvectors[0, 0], eigenvectors[0, 1], head_width=0.1, head_length=0.1, color='r', label='Eigenvector 1')
plt.arrow(0, 0, eigenvectors[1, 0], eigenvectors[1, 1], head_width=0.1, head_length=0.1, color='b', label='Eigenvector 2')
plt.axis('equal')
plt.title("Eigenvector Visualization")
plt.legend()
plt.show()
```
